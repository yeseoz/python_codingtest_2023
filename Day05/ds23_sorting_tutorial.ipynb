### 선택 정렬

#### 최솟값 위치 확인 함수

```python
def findMinIdx(ary):
    minIdx = 0
    for i in range(1, len(ary)):
        if (ary[minIdx] > ary[i]): # ary [i] 보다 ary[mminIdx]가 크면
            minIdx = i # minIdx를 작은 값으로 바꿔준다
    return minIdx
```

```python
tests = [55, 88, 33, 77]
minPos = findMinIdx(tests)
print('최솟값 --> ', tests[minPos])
```

#### 선택 정렬 전체

```python
# 변수선언
before = [188, 162, 168, 120, 50, 150, 177, 105]
after = []

print('정렬전 -->', before)
for _ in range(len(before)):
    minPos = findMinIdx(before)
    after.append(before[minPos])
    del(before[minPos]) # before에서 찾은 최소값 지워줌

print('정렬 후 --> ', after)
```

#### 삽입정렬 Insertion Sort

##### 자신의 삽입위치를 찾는 함수

```python
def findInsertIdx(ary, data):
    findIdx = -1
    for i in range(0, len(ary)):
        if (ary[i] > data):
            findIdx = i
            break
    if findIdx == -1:
        return len(ary)
    else: 
        return findIdx
    
# 큰수가 나오면 무조건 그 앞에 들어가게 됨....
```

```python
testAry = []
print('55가 들어갈 위치 --> ', findInsertIdx(testAry, 50))

testAry = [33, 77, 88]
print('55가 들어갈 위치 --> ', findInsertIdx(testAry, 55))
```

#### 삽입정렬 개선

```python
def insertionSort(ary):
    n = len(ary)
    for end in range(1, n):
        for cur in range(end, 0, -1):
            if(ary[cur-1] > ary[cur]):
                # tmp = ary[cur-1]
                # ary[cur-1] = ary[cur]
                # ary[cur] = tmp
                ary[cur-1], ary[cur] = ary[cur], ary[cur-1] #???!! => 파이썬 swap
    return ary

# ㅈ전역변수 선언 부분
dataAry = [188, 162, 168, 120, 50, 150, 177, 105]

# 메인코드 부분
print('정렬전 --> ', dataAry)
dataAry = insertionSort(dataBank)
print('정렬후 --> ', dataAry)
```

#### 버블 정렬 (Bubble Sort)

```python
def BubbleSort(ary):
    n = len(ary)
    for end in range(n-1, 0, -1): # n-1에서 0 까지 역순으로 진행
        #print('사이클# --> ', ary)
        for cur in range(0, end): # 0에서 n-2까지 반복
            if(ary[cur] > ary[cur+1]):
                ary[cur], ary[cur+1] = ary[cur+1], ary[cur]
    #return ary
        
# 전역 변수 선언 부분
dataAry = [188, 162, 168, 120, 50, 150, 177, 105]

# 메인코드
print('정렬전 --> ', dataAry)
BubbleSort(dataBank)
print('정렬 후 -->', dataBank)
```

#### 퀵 정렬(Quick Sort)

```python
def quickSort(ary):
    n = len(ary)
    if n <= 1:
        return ary
    
    pivot = ary[n//2]  # ary[4]
    leftAry, rightAry, midAry = [], [], []

    for num in ary:
        if num < pivot:
            leftAry.append(num)
        elif num > pivot:
            rightAry.append(num)
        else:
            midAry.append(num) # 중복을 고려

    return quickSort(leftAry) + [pivot] + quickSort(rightAry)


# 전역 변수 선언 부분
dataAry = [188, 162, 168, 120, 50, 150, 177, 105]

# 메인코드
print('정렬전 --> ', dataAry)
dataAry = quickSort(dataBank)
print('정렬 후 -->', dataBank)
```
